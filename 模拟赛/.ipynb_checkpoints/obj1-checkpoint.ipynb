### B：行动 5’

【问题描述】

小明站在坐标（0, 0）处，面朝 x轴正方向。
第一轮，他向前走 1单位距离，然后右转；
第二轮，他向前走 2单位距离，然后右转；
第三轮，他向前走 3单位距离，然后右转……他一直这么走下去。
请问第 2020轮后，他的坐标是：（_________，_________）。

```python
polt = [0, 0]

flag = 0
for i in range(1, 2020+1):
    if flag > 3:
        flag = 0 
    
    if flag == 0:
        polt[0] = polt[0] + i
        flag = flag+1
        continue
        
    elif flag == 1:
        polt[1] = polt[1] - i
        flag = flag+1
        continue
        
    elif flag == 2:
        polt[0] = polt[0] - i
        flag = flag+1
        continue
        
    elif flag == 3:
        polt[1] = polt[1] + i
        flag = flag+1
        continue
print(*polt)
```

取反操作，python取反的是按补码取反，即是源码加1再取反
flag = (~flag)+1

```python
result = 0

flag = 1
for i in range(1, 4040, 2):
    result = result +  flag * 4 * 1/i 
    flag = (~flag)+1
print('%.6f' % result)
```

### 分解质因数

```python
res = ''

def solve(res,n):
    
    for i in range(2, n+1):
        
        if n % i ==0:
            
            res += str(i)
            
            n = n // i
            
            if n == 1:
                
                return res
            
            else:
                
                res += '*'
                
                return solve(res,n)
        else:
            continue
            
print(solve(res,2020))
```

### F：你好，2020

2020年，这个年份很特别，2020从中间分成两个整数，大小形状完全一样。

小明对形如2020的数字很感兴趣(不包括前导零)，在1到1200中这样的数字包括11、22、33、44、55、66、77、88、99、1010、1111共11个，他们的和是2616

请问，在1到n中，所有这样的数的和是多少？

```python
n = int(input())

result = 0

for i in range(1, n+1):
    string = str(i)
    length = len(string) 
    if  length % 2 == 0:
        if string[:length//2] == string[length//2:]:
            
            result = result + i
        
print(result)
```

### 组合搭配

```python
n = int(input())

chars = list()

for _ in range(n):
    chars.extend(list(map(str, input().split())))
    
def calute(chars):
    result = 0
    def value():
    
        return (ord(char[0].upper()) - 64) * len(char) * chars.index(char) 
    
    for char in chars:
        
        result = result + value()

    return result

from itertools import product

chars = list(product(chars, repeat=n))

_chars = chars.copy()

for value in chars:
    if len(set(chars)) != n:
        _chars.remove(value)
        
        
print(_chars)
```

### J：因数个数

描述
求所有 2 到 n 的整数中，因数个数第k少的数因数个数是多少。

输入
第一行两个正整数 n,k即题目描述中的 n,k

输出
输出仅一行，即因数个数第 k少的数因数个数。

```python
def solve(res,n):
    
    for i in range(2, n+1):
        
        if n % i ==0:
            
            res.append(i)
            
            n = n // i
            
            if n == 1:
                
                return res
            
            else:
                    
                return solve(res,n)
            
        else:
            continue

n, k = map(int, input().split())

results = list()

for i in range(2, n+1):
    results.append(list())
    
    solve(results[-1], i)


results = [len(result) for result in results]

results.sort()

print(results[k])
```
