```python
"""
上午游玩的游客要求按照预 约的顺序排列后，名字是单调递增的，即排在前面的名字严格小于排在后面的 名字。
一个名字 A 小于另一个名字 B 是指：存在一个整数 i，使得 A 的前 i 个字母与 B 的前 i 个字母相同，
且 A 的第i+1 个字母小于 B 的第i+1 个字母。（如果 A 不存在第 i+1 个字母且 B 存在第 i+1 个字母，
也视为 AA 的第 i+1 个字母小于 B 的第 i + 1 i+1 i+1 个字母）
作为小蓝的助手，你要按照小蓝的想法安排游客，同时你又希望上午有尽 量多的游客游玩，
请告诉小蓝让哪些游客上午游玩。如果方案有多种，请输出 上午游玩的第一个游客名字最小的方案。
如果此时还有多种方案，请输出第一 个游客名字最小的前提下第二个游客名字最小的方案。如果仍然有多种，
依此 类推选择第三个、第四个……游客名字最小的方案。
"""
```

```python
# 每个游客的名字由一个大写英文字母开始，后面跟 0 个或多个小写英文字母。游客可能重名

string = input()

names = list(string)
for i in range(1, len(string)):
    if string[i].isupper():
        names.insert(names.index(string[i])," ")
        
names = "".join(names).split(" ")

ords = [[ord(char) for char in name] for name in names]

_ords = list()

Flag = True
i = 0
while Flag:
    re = list()
    for temp in ords:
        if len(temp) > i:
            re.append(temp[i])
            Flag = True
        else:
            Flag = False
    
    i = i+1
    _ords.append(re)
    
print(names)
print(_ords)

if len(set(_ords[0])) == len(_ords[0]):
    longer_lis = lis(_ords[0])
    for num in longer_lis:
        print(names[_ords[0].index(num)])
        
```

```python
# 最长递增子序列
def lis(arr):
    n = len(arr)
    m = [0]*n
    for x in range(n-2, -1, -1):
        for y in range(n-1, x, -1):
            if arr[x] < arr[y] and m[x] <= m[y]:
                m[x] += 1
        
        max_value = max(m)
        
        result = []
        for i in range(n):
            if m[i] == max_value:
                result.append(arr[i])
                max_value -= 1
    
    return result
```

```python
def LongestOfLIS(nums):
    length = len(nums)
    dp = [0]*length
    
    for i in range(length-2, -1, -1):
        for j in range(length-1, i, -1):
            if nums[i] < nums[j] and dp[i] <= dp[j]:
                dp[i] = dp[i] + 1
                
        max_len = max(dp)
        
    
        # dp[i] 表示包含序号i元素的最长上升子序列的长度
        result = list()
        for i in range(length):
            if dp[i] == max_len:
                result.append(nums[i])
                max_len = max_len - 1
                
    return result
    
    
    """
    return max_len
    """
```

```python
def lengthOfLIS(nums):
        max_len = 1
        nums_len = len(nums)
        dp = [1] * nums_len  # dp[i] 表示包含序号i元素的最长上升子序列的长度
        for i in range(1,nums_len):
            for j in range(i):
                if nums[j]<nums[i]:
                    dp[i] = max(dp[i],dp[j]+1)
            
            max_len = max(max_len, dp[i])
        
        return max_len
```
