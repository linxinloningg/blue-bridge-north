```python
def judge(num):
    import math
    if num <= 1:
        return False
    else:
        for i in range(2, int(math.sqrt(num))+1):
            if num % i ==0:
                return False
        
        return True

s = list()
for i in range(2, 2019):
    if judge(i):
        s.append(i)

import itertools

combinations = list(itertools.combinations(s, 2))

for combination in combinations:
    if sum(combination) == 2019:
        print(combination)
```

```python
试题 D: 求值
【问题描述】
学习了约数后，小明对于约数很好奇，他发现，给定一个正整数 t，总是可
以找到含有 t 个约数的整数。小明对于含有 t 个约数的最小数非常感兴趣，并
把它定义为 S t 。
例如 S 1 = 1, S 2 = 2, S 3 = 4, S 4 = 6，· · · 。
现在小明想知道，当 t = 100 时，S t 是多少？即 S 100 是多少？
```

```python
num = 1
while True:
    counter = 0
    for i in range(1, num+1):
        if num % i == 0:
            counter = counter + 1
    if counter == 100:
        break
    else:
        num += 1

print(num)
```

```python
S = input()
T = input()

index = 0
num = 0
for char in T:
    for i in range(index, len(S)):
        if S[i] == char:
            index = i + 1
            break
    else:
        if index == 0:
            index = 1
        num = num + 1

print(num)
```

```python
def judge(nums, k):
    counter = 1
    for i in range(1, len(nums)-1):
        if nums[i-1] > nums[i] < nums[i+1] or nums[i-1] < nums[i] > nums[i+1]:
            counter = counter + 1
    if counter == k:
        return True
    else:
        return False

n, k = map(int, input().split())

import itertools
nums_list = list(itertools.permutations(list(range(1, n+1)), n))

result = 0
for nums in nums_list:
    if judge(nums, k):
        result = result + 1
        
print(result)
```
