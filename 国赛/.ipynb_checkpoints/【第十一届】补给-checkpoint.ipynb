```python
n, D = map(int, input().split())

spots = [list(map(int, input().split()))for i in range(n)]

_ = spots.pop(0)
origin = _

import math

results = list()

while len(spots) > 0:
    for spot in spots:
        distant = math.sqrt(pow(spot[0]-origin[0], 2)+pow(spot[1]-origin[1], 2))
        if  distant > D:
            continue
        else:
            results.append(distant)
            origin = spot
            spots.remove(spot)
            
print(sum(results)*2)
```
