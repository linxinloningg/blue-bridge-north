```python
import itertools
# 从左到右依次为：47 46 61
sum_num = 154
num = [2,5,9,11,16,17,19,21,22,24,26,30,31,33,35,36,41,50,52]

left = list()
right = list()
bottom = list()
for i in range(2, len(num)):
    com_list = list(itertools.combinations(num, i))
    for com in com_list:
        if sum(com) + 47 == sum_num:
            left.append(com)
        if sum(com) + 61 == sum_num:
            right.append(com)
        if sum(com) == sum_num:
            bottom.append(com)
```
