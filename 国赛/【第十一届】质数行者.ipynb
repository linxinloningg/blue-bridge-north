### 判断质数

```python
def isPrime(num):
    import math
    if num <= 1:
        return False
    else:
        for i in range(2, int(math.sqrt(num))+1):
            if num % i == 0:
                return False
            
        return True
```

```python
n, m, w = map(int, input().split())

r1 , c1 , h1 , r2 , c2 , h2 = map(int, input().split())

x = list()
y = list()
z = list()
for i in range(n+1):
    if isPrime(i):
        x.append(i)
for i in range(1, m+1):
    if isPrime(i):
        y.append(i)
for i in range(1, w+1):
    if isPrime(i):
        z.append(i)
        
import itertools
results = list()
for i in range(len(x)):
    per_res = list(itertools.combinations(x, i))
    for re in per_res:
        if sum(re) == n:
            if len(re) > 1:
                results.extend(list(itertools.permutations(re, len(re))))
            else:
                results.append(re)

x = results

results = list()
for i in range(len(y)):
    per_res = list(itertools.combinations(y, i))
    for re in per_res:
        if sum(re) == m:
            if len(re) > 1:
                results.extend(list(itertools.permutations(re, len(re))))
            else:
                results.append(re)

y = results

results = list()
for i in range(len(z)):
    per_res = list(itertools.combinations(z, i))
    for re in per_res:
        if sum(re) == w:
            if len(re) > 1:
                results.extend(list(itertools.permutations(re, len(re))))
            else:
                results.append(re)
            
z = results

product_res = list(itertools.product(x, y))
print(len(product_res))
```
