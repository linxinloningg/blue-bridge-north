```python
import math

def judge(num):
    if num == 1:
        return False
    else:
        for i in range(2, int(math.sqrt(num)) + 1):
            if num % i == 0:
                return True
            
    return False
    
result = 0
for i in range(1, 2021):
    if judge(i):
        result = result + 1
        
print(result)
```

```python
def expand(spot):
    
    up = (spot[0], spot[0]+1)
    down = (spot[0], spot[0]-1)
    left = (spot[0]-1, spot[0])
    right = (spot[0]+1, spot[0])
    return [up, down, left, right]


spots = [[(0,0),(2020,11),(11,14),(2000,2000)]]

for i in range(2020):
    
    expand_spot = list()
    for spot in spots[i]:
        expand_spot.extend(expand(spot))
        
    spots.append(expand_spot)


spots = set([spot for spot_list in spots for spot in spot_list ])

print(len(spots))
```

```python
string = list("tocyj")

def judge(son_string, string):
    # 在一个字符串中，如果取出若干个字符，将这些字符按照在字符串中的顺序排列后是单调递增的
    # 则成为这个字符串中的一个单调递增子序列
    for i in range(len(son_string)-1):
        if string.index(son_string[i]) >= string.index(son_string[i+1]):
            return False
    
    return True

result = list()
import itertools
for i in range(2, 5):
    son_list = list(itertools.combinations(string, i))
    for son_string in son_list:
        if judge(son_string, string):
            result.append(son_string)
            
result = set(result)
print(len(result) + 5)
```
