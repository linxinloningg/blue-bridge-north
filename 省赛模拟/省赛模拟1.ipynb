```python
import itertools
data = ['(',')','(',')','(',')','(',')','(',')','(',')','(',')','(',')'] 
product_res = itertools.product(data, data)

num = 0
for part in list(product_res):
    if part == part[::-1]:
        num += 1

print(num)    
```

### 编程题1 凯撒密码加密

```python
string = list(input())

for i in range(len(string)):
    
    char_ord = ord(string[i]) + 3
        
    string[i] = chr(char_ord if char_ord <= 122 else char_ord - 26)

print("".join(string))
```

### 编程题2 反倍数

```python
n = int(input())

a, b, c = map(int, input().split())

nums = list()

for i in range(1, n+1):
    
    if i % a != 0 and i % b != 0 and i % c != 0:
        nums.append(i)
        
print(len(nums))
```

### 编程题3 摆动序列

```python
m, n = map(int, input().split())

num = [i for i in range(1, n+1)]

from itertools import product

nums = list(product(num, repeat=3))


def judge(value):
    
    for i in range(2, len(value), 2):
        
        if value[i] <= value[i-1]:
            
            return False
        
        
    for i in range(1, len(value), 2):
        
        if value[i] >= value[i-1]:
            
            return False
        
    return True
    
for value in nums:
    
    if judge(value):
        
        print(value)
```

### 编程题4 螺旋矩阵

```python
n, m = map(int, input().split())

r, c = map(int, input().split())

nums = [[0 for _ in range(m)] for _ in range(n)]
_nums = [[0 for _ in range(m)] for _ in range(n)]

x, y = 0, 0

i = 1
while i < n * m:
    while y < m and _nums[x][y] == 0:
        nums[x][y] = i
        _nums[x][y] = 1
        i = i + 1
        y = y + 1
        
    y -= 1
    x += 1
    
    while x < n and _nums[x][y] == 0:
        nums[x][y] = i
        _nums[x][y] = 1
        i = i + 1
        x = x + 1
        
    x -= 1
    y -= 1
    
    while y >= 0 and _nums[x][y] == 0:
        nums[x][y] = i
        _nums[x][y] = 1
        i = i + 1
        y = y - 1
        
    x -= 1
    y += 1
    
    while x >= 0 and _nums[x][y] == 0:
        nums[x][y] = i
        _nums[x][y] = 1
        i = i + 1
        x = x - 1
        
    x += 1
    y += 1
    

print(nums)
```

### 编程题5 村庄通电

```python
from math import sqrt
n = int(input())

countrys = [list(map(int, input().split())) for _ in range(n)]

center = countrys.pop(0)

def calculate(center, countrys):
    
    des = [pow(country[0]-center[0], 2)+
           pow(country[1]-center[1], 2)+
           pow(country[2]-center[2], 2) for country in countrys]
    
    country_index = des.index(min(des))
    
    return country_index

result = 0

while len(countrys) != 0:
    
    country_index = calculate(center, countrys)
    
    result = result + sqrt(pow(center[0] - countrys[country_index][0], 2)+ 
                           pow(center[1] - countrys[country_index][1], 2)+
                           pow(center[2] - countrys[country_index][2], 2))
                            
    center = countrys.pop(country_index)


print(round(result, 2))
```
