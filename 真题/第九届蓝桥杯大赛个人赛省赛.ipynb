### 递增三元组 

```python
N = int(input())

nums = [list(map(int, input().split())) for _ in range(N)]

from itertools import product

production = list(product(*nums))

num = 0
for _list in  production:
    i = _list[0]
    j = _list[1]
    k = _list[2]
    
    if 1 <= i <= N and 1 <= j <= N and 1 <= k <= N and i <= j <= k:
        num += 1
        
print(num)
```

### 全球变暖

```python
N = int(input())

photo = [list(input()) for _ in range(N)]

tmp = photo.copy()
for i in range(1, N-1):
    for j in range(1, N-1):
        print(tmp[i][j])
        if tmp[i-1][j] == '#' and tmp[i+1][j] == '#' and tmp[i][j-1] == '#' and tmp[i][j+1] == '#':
            pass
        else:
            photo[i][j] = '.'

photo = ["".join(line) for line in photo]
print("\n\n")
print(*photo,sep='\n')
```

### 乘积最大

```python
N, K = map(int, input().split())

nums = [int(input()) for _ in range(N)]

chose = list()
for _ in range(K):
    max_num = max(nums)
    chose.append(max_num)
    nums.remove(max_num)

    
from math import  prod
print(prod(chose) % 1000000009)
```
