```python
with open('test.txt', 'r') as f:
    content = f.read()

content = content.split('\n')[:-1]

count = 0
for line in content:
    count += line.count('2020')

content = list(map(list, zip(*content)))

content = ["".join(line) for line in content]

for line in content:
    count += line.count('2020')

print(count)
```

### 跑步

```python
from datetime import datetime, timedelta
start = datetime.strptime('2000/01/01', '%Y/%m/%d')
end = datetime.strptime('2020/10/01', '%Y/%m/%d')

num = 0
today = start
while True:
    if today == end:
        break
    
    if today.day == 1:
        num += 2
    elif today.weekday() == 0:
        num += 2
    else:
        num += 1
    today = today + timedelta(1)
    
print(num)
```

### 成绩统计

```python
n = int(input())
nums = [int(input()) for _ in range(n)]

ok_rate = 0
prefect_rate = 0
for num in nums:
    if num >= 60:
        ok_rate += 1
    if num >= 85:
        prefect_rate += 1
print(ok_rate/n)
print(prefect_rate/n)
```

### 数字三角形

```python
n = int(input())

nums = [list(map(int, input().split())) for _ in range(n)]

dp = nums.copy()

for i in range(1, n):
    for j in range(i+1):
        if j == 0:
            dp[i][j] += dp[i-1][j]
        elif j == i:
            dp[i][j] += dp[i-1][j-1]
        else:
            dp[i][j] += max(dp[i-1][j],dp[i-1][j-1])
if n&1:
    print(dp[n-1][n//2])
else:
    print(max(dp[n-1][n//2-1],dp[n-1][n//2]))
```

### 平面划分

```python
n = int(input())

lines = [tuple(map(int, input().split())) for _ in range(n)]

lines = list(set(lines))

from itertools import combinations
lines_combination = list(combinations(lines, 2))

pots = 0
for combination in lines_combination:
    if combination[0][0] != combination[1][0]:
        pots += 1
        
print(n+pots+1)
```
