### 组队

```python
with open('team.txt', 'r') as f:
    content = f.read()
    
nums = content.split('\n')[:-1]

nums = [line.split(' ') for line in nums]

nums = [[int(num) for num in line] for line in nums]

nums = [sum(line) for line in nums]

peoples = list()
for i in range(5):
    people = max(nums)
    peoples.append(people)
    nums.remove(people)
    
print(sum(peoples))
```

### 年号字串

```python
g = 2019 % 26
print(g)
ans = 2019 - g
while True:
    ans = ans // 26
    if ans > 26:
        print(ans % 26)
    else:
        print(ans)
        break
```

### 特别数的和

```python
arr = ['2', '0', '1', '9']

n = int(input())

results = list()
for i in range(1, n+1):
    string = list(str(i))
    for char in string:
        if char in arr:
            results.append(i)
            break
            
print(sum(results))
```

### 完全二叉树的权值

```python
N = int(input())

nums = list(map(int, input().split()))

i = 1
count = list()
while True:
    count.append(i)
    i *= 2
    if sum(count) > N:
        break

tree = list()

for value in count:
    
    tmp = nums[:value]
    tree.append(tmp)
    for num in tmp:
        nums.remove(num)
    
results = [sum(col) for col in tree]

print(results.index(max(results)) + 1)
```

### 等差数列

```python
N = int(input())

nums = list(map(int, input().split()))

nums.sort()

sep = 0
for i in range(len(nums) - 2):
    if nums[i] - nums[i+1] == nums[i+1] - nums[i+2]:
        sep = nums[i+1] - nums[i]
        break

results = list()
for value in range(nums[0], nums[-1]+1, sep):
    results.append(value)
    
print(len(results))
```

### 后缀表达式

```python
N, M = map(int, input().split())

nums = list(map(int, input().split()))

nums.sort(reverse=True)
count = 0
result = nums[0]
for num in nums[1:]:    
    if count < N:
        result += num
    else:
        result -= num
    count += 1
    
print(result)
```
