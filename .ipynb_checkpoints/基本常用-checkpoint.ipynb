### 常用标准库

```python
math
数论
math.ceil(3.5)  # 4 向上取整
math.floor(3.5)  # 3 向下取整
math.fabs(-3)  # 3.0 绝对值
math.factorial(4)  # 24 阶乘
math.comb(4, 2)  # 6(组合数)
math.perm(4, 2)  # 12（排列数）
math.isqrt(5)  # 2(根号5向下取整)
math.gcd(4, 6)  # 2(最大公约数)
math.lcm(4, 6)  # 12(最小公倍数)
math.prod([2, 3, 5])  # 30 乘积
math.isfinite(x)  # x既不是无穷大也不是NaN返回True
math.isinf(x)  # x是正或负无穷大，返回True
math.isnan(x)  # x是NaN，返回True
```

```python
数学函数(结果均为浮点数)
math.exp(x)          # 返回 e 次 x 幂
math.expm1(x)        # 返回 e 的 x 次幂，减1
math.log(x[, base])  
# 使用一个参数，返回 x 的自然对数（底为 e ）。
# 使用两个参数，返回给定的 base 的对数 x ，计算为 log(x)/log(base)
math.pow(x, y)  # x的y次方
math.sqrt(x)    # x 的平方根
math.sin(x)
math.cos(x)
math.tan(x)
```

### collections

```python
collections.Counter('aabbccc')  # Counter({'c': 3, 'a': 2, 'b': 2})
collections.Counter(['a', 'a', 'b'])  # Counter({'a': 2, 'b': 1})
collections.Counter(cats=4, dogs=8)  # Counter({'dogs': 8, 'cats': 4})
```

### elements

```python
import collections
data = collections.Counter(a=4, b=2, c=0, d=-2)

data = list(data.elements())
print(data)  # ['a', 'a', 'a', 'a', 'b', 'b']
```

### most_common

```python
import collections
data = collections.Counter('abracadabra')
print(data.most_common(3))  # [('a', 5), ('b', 2), ('r', 2)]
```

### subtract

```python
从 迭代对象 或 映射对象 减去元素。输入和输出都可以是0或者负数。
c = Counter(a=4, b=2, c=0, d=-2)
d = Counter(a=1, b=2, c=3, d=4)
print(c.subtract(d))  # Counter({'a': 3, 'b': 0, 'c': -3, 'd': -6})
```

### 列表转置

```python
列表的"*"操作

在迭代对象前使用‘*’号，将变量中可迭代对象的元素拆解出来，变成若干独立的参数。

_A = list(map(list, zip(*A))) # list转置

zip() 函数用于将可迭代的对象作为参数，将对象中对应的元素打包成一个个元组，然后返回由这些元组组成的列表。

zip 方法在 Python 2 和 Python 3 中的不同：在 Python 3.x 中为了减少内存，zip() 返回的是一个对象。如需展示列表，需手动 list() 转换。

map() 会根据提供的函数对指定序列做映射.
第一个参数 function 以参数序列中的每一个元素调用 function 函数，返回包含每次 function 函数返回值的新列表。
```

### 排列组合

```python
 import itertools
 data = [3, 4, 6]

 com_res = itertools.combinations(data, 2)
 print(list(com_res))
 # [(3, 4), (3, 6), (4, 6)]

 per_res = itertools.permutations(data, 2)
 print(list(per_res))
 # [(3, 4), (3, 6), (4, 3), (4, 6), (6, 3), (6, 4)]


 com_without_replacement_res = itertools.combinations_with_replacement(data, 2)
 print(list(com_without_replacement_res))
 # [(3, 3), (3, 4), (3, 6), (4, 4), (4, 6), (6, 6)]

 product_res = itertools.product(data, data)
 print(list(product_res))
 # [(3, 3), (3, 4), (3, 6), (4, 3), (4, 4), (4, 6), (6, 3), (6, 4), (6, 6)]
```

```python
combinations 组合
permutations 排列

combinations_with_replacement 组合可重复
product 排列可重复
```
