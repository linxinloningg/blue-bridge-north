```python
# 2元的张数刚好是1元的张数的10倍

i = 1
while True:
    if (200 - (i + 10 * i * 2)) % 5 == 0 :
        break
    else:
        i = i + 1
        
print(i)
```

```python
# 相邻的两台激光器不能同时打开

n = 30
result = 0
for i in range(n):
    for j in range(n, 0, -(n-i)):
        result = result + 1

print(result-1)
```

```python
# 在 M78 星云，时间的计量单位和地球上不同，M78 星云的一个小时有 n 分钟
# 表示把当前的数加 k

n, k = map(int, input().split())

i = 1
while True:
    if  (i * k) % n == 1:
        break
    else:
        i = i + 1

print(i)
```

```python
n, m = map(int, input().split())

spots = [list(input()) for i in range(n)]


# 规则1：每块积木必须紧挨着放置在某一块积木的正上方，与其下一层的积木对齐；
# 规则2：同一层中的积木必须连续摆放，中间不能留有空隙；
# 规则3：小明不喜欢的位置不能放置积木

spots = list(map(list, zip(*spots)))

result = 1
for spot_list in spots:
    for spot in spot_list[::-1]:
        if spot == '.':
            result = result + 1

print(result)
```

```python
import math

n = int(input())

result = 0
for r in range(1, n+1):
    row = list()
    for c in range(1, n+1):
        row.append(pow(math.gcd(r, c), 2))
    result = result + sum(row)
    
print(result)
```
